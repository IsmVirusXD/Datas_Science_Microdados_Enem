```python
#Dicionario para converter os Numeros em Texto
dictraca = {
  0	: "Não declarado",
  1	: "Branca",
  2	: "Preta",
  3	: "Parda",
  4	: "Amarela",
  5	: "Indígena" 
}

#Filtros Genericos
notas = ['NU_NOTA_MT','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_CN','NU_NOTA_REDACAO']
presenca = ['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','TP_PRESENCA_MT']
```

###Analise 2017

```python
filtro_ausente    = ((dp_Enem2017['TP_PRESENCA_CN'] == 0) | (dp_Enem2017['TP_PRESENCA_CH'] == 0) | (dp_Enem2017['TP_PRESENCA_LC'] == 0) | (dp_Enem2017['TP_PRESENCA_MT'] == 0))
filtro_presente   = ((dp_Enem2017['TP_PRESENCA_CN'] == 1) & (dp_Enem2017['TP_PRESENCA_CH'] == 1) & (dp_Enem2017['TP_PRESENCA_LC'] == 1) & (dp_Enem2017['TP_PRESENCA_MT'] == 1))
filtro_eliminado  = ((dp_Enem2017['TP_PRESENCA_CN'] == 2) | (dp_Enem2017['TP_PRESENCA_CH'] == 2) | (dp_Enem2017['TP_PRESENCA_LC'] == 2) | (dp_Enem2017['TP_PRESENCA_MT'] == 2))
```

```python
# Dividindo o Banco de Dadoo:
selecao_linhas = ['NU_INSCRICAO']
ausentes    = dp_Enem2017.loc[filtro_ausente,selecao_linhas]

eliminados  = dp_Enem2017.loc[filtro_eliminado,selecao_linhas]

presentes   = dp_Enem2017.loc[filtro_presente,selecao_linhas]

au = ausentes.count()
el = eliminados.count()
pr = presentes.count()

total = au + el + pr

print("Ausentes " + str(au))
print("Eliminados " + str(el))
print("Presentes " + str(pr))
```

```python
percent = total 
percent
```

```python
# Dividindo o Banco de Dadoo:
selecao_linhas = ['NU_INSCRICAO','TP_COR_RACA']
ausentes    = dp_Enem2017.loc[filtro_ausente,selecao_linhas]
ausentes['COR/RACA'] = [dictraca[valor] for valor in ausentes['TP_COR_RACA']] # <- Criando uma coluna nova para ajudar na identificação de alguns valores

eliminados  = dp_Enem2017.loc[filtro_eliminado,selecao_linhas]
eliminados['COR/RACA'] = [dictraca[valor] for valor in eliminados['TP_COR_RACA']]

selecao_linhas = ['NU_INSCRICAO','TP_COR_RACA'] + notas
presentes   = dp_Enem2017.loc[filtro_presente,selecao_linhas]
presentes['COR/RACA'] = [dictraca[valor] for valor in presentes['TP_COR_RACA']]
```

```python
#Fazendo a Contagem do Numero de Ausentes
filtro = ['COR/RACA','NU_INSCRICAO',]
ausentes.filter(filtro).groupby('COR/RACA').count().sort_values(by=['NU_INSCRICAO'],ascending=False)
```

```python
#Contegam de Pessoas Eliminadas
filtro = ['COR/RACA','NU_INSCRICAO']
eliminados.filter(filtro).groupby('COR/RACA').count().sort_values(by=['NU_INSCRICAO'],ascending=False)
```

```python
# Media das Notas de Cada Prova
filtro = ['COR/RACA'] + notas
presentes.filter(filtro).groupby('COR/RACA').mean().sort_values(by=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT'],ascending=False)
```

```python
#Valores Maximos para Cada Prova
presentes.filter(filtro).groupby('COR/RACA').max().sort_values(by=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT'],ascending=False)
```

```python
#Valor Minimo da Prova de MT
selecao_linhas = ['NU_INSCRICAO','TP_COR_RACA','COR/RACA'] + notas
minimos = presentes.loc[(presentes['NU_NOTA_MT'] != 0),selecao_linhas]
filtro = ['COR/RACA','NU_NOTA_MT'] 
minimos.filter(filtro).groupby('COR/RACA').min().sort_values(by=['NU_NOTA_MT'],ascending=False)
```

```python
#Valor Minimo da Prova de LC
minimos = presentes.loc[(presentes['NU_NOTA_LC'] != 0),selecao_linhas]
filtro = ['COR/RACA','NU_NOTA_LC'] 
minimos.filter(filtro).groupby('COR/RACA').min().sort_values(by=['NU_NOTA_LC'],ascending=False)
```

```python
#Valor Minimo da Prova de CN
minimos = presentes.loc[(presentes['NU_NOTA_CN'] != 0),selecao_linhas]
filtro = ['COR/RACA','NU_NOTA_CN'] 
minimos.filter(filtro).groupby('COR/RACA').min().sort_values(by=['NU_NOTA_CN'],ascending=False)
```

```python
##Valor Minimo da Prova de CH
minimos = presentes.loc[(presentes['NU_NOTA_CH'] != 0),selecao_linhas]
filtro = ['COR/RACA','NU_NOTA_CH'] 
minimos.filter(filtro).groupby('COR/RACA').min().sort_values(by=['NU_NOTA_CH'],ascending=False)
```

```python
#Criando a Tabela de Soluções
df_resultado = pd.DataFrame({'Raca/Cor': ["Não declarado","Branca","Preta","Parda","Amarela","Indígena"]})
df_resultado.head()

#Adicionando Ausentes
novo = ausentes.groupby('TP_COR_RACA').count().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Ausentes'] = novo.loc[:,'NU_INSCRICAO']

#Adicionando Presentes
novo =  presentes.groupby('TP_COR_RACA').count().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Presentes'] = novo.loc[:,'NU_INSCRICAO']

#Adicionando Media dos Candidatos
filtro = ['TP_COR_RACA'] + notas
novo = presentes.filter(filtro).groupby('TP_COR_RACA').mean().sort_values(by=['TP_COR_RACA'])
df_resultado['Med_CN'] = novo.loc[:,'NU_NOTA_CN']      # Ciencias Naturais
df_resultado['Med_CH'] = novo.loc[:,'NU_NOTA_CH']      # Ciencias Humanas
df_resultado['Med_LC'] = novo.loc[:,'NU_NOTA_LC']      # Linguagens 
df_resultado['Med_MT'] = novo.loc[:,'NU_NOTA_MT']      # Matematica
df_resultado['Med_RD'] = novo.loc[:,'NU_NOTA_REDACAO'] # Redação

#Adicionando as Notas Mais Altas
novo = presentes.filter(filtro).groupby('TP_COR_RACA').max().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Max_CN'] = novo.loc[:,'NU_NOTA_CN']      # Ciencias Naturais
df_resultado['Max_CH'] = novo.loc[:,'NU_NOTA_CH']      # Ciencias Humanas
df_resultado['Max_LC'] = novo.loc[:,'NU_NOTA_LC']      # Linguagens
df_resultado['Max_MT'] = novo.loc[:,'NU_NOTA_MT']      # Matematica
df_resultado['Max_RD'] = novo.loc[:,'NU_NOTA_REDACAO'] # Redação

#Adicionando as Menores Notas não nulas
minimos = presentes.loc[(presentes['NU_NOTA_CN'] != 0),selecao_linhas] # Ciencias Naturais
novo = minimos.groupby('TP_COR_RACA').min().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Min_CN'] = novo.loc[:, 'NU_NOTA_CN']

minimos = presentes.loc[(presentes['NU_NOTA_CH'] != 0),selecao_linhas] # Ciencias Humanas
novo = minimos.groupby('TP_COR_RACA').min().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Min_CH'] = novo.loc[:, 'NU_NOTA_CH']

minimos = presentes.loc[(presentes['NU_NOTA_LC'] != 0),selecao_linhas] # Linguagem
novo = minimos.groupby('TP_COR_RACA').min().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Min_LC'] = novo.loc[:,'NU_NOTA_LC']

minimos = presentes.loc[(presentes['NU_NOTA_MT'] != 0),selecao_linhas] # Matematica
novo = minimos.groupby('TP_COR_RACA').min().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Min_MT'] = novo.loc[:,'NU_NOTA_MT'] 

minimos = presentes.loc[(presentes['NU_NOTA_REDACAO'] != 0),selecao_linhas] # Redação
novo = minimos.groupby('TP_COR_RACA').min().sort_values(by=['TP_COR_RACA'],ascending=True)
df_resultado['Min_RD'] = novo.loc[:,'NU_NOTA_REDACAO']                     
```
